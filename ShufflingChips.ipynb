```python
def shuffle_once(left_arr, right_arr):

    assert(len(left_arr) == len(right_arr))
    
    # by default shift left above right
    new_arr = []
    for i in range(len(left_arr)):
        new_arr.append(right_arr[i])
        new_arr.append(left_arr[i])

    return new_arr

def split(arr):

    assert(len(arr) % 2 == 0)
    return arr[:int(len(arr)/2)], arr[int(len(arr)/2):] # top half goes to the right

def shuffles_to_return(larr, rarr):

    new_l, new_r = split(shuffle_once(larr, rarr))
    counter = 1
    while ((new_l != larr) and (new_r != rarr)):
        new_l, new_r = split(shuffle_once(new_l, new_r))
        counter += 1

    return counter
    
```

```python
import matplotlib.pyplot as plt

# plot stack height vs STR
xvals=[]
yvals=[]
for i in range(1, 501):
    larr = [1] * i
    rarr = [0] * i
    xvals.append(i)
    yvals.append(shuffles_to_return(larr, rarr))

plt.plot(xvals, yvals)
plt.title("Stack Height vs STR")
plt.xlabel("Height of Each Stack")
plt.ylabel("Number of Shuffles to Return to Original State")
plt.show()
```
