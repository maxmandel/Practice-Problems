```python
import numpy as np
import matplotlib.pyplot as plt

die_outcomes = [1, 2, 3, 4, 5, 6]
min_payoffs = [0]*15
xvals = [0]*15
yvals = [1]*15
for i in range(1, 15):
    total = 0
    for outcome in die_outcomes:
        if outcome > min_payoffs[i-1]:
            total += outcome
        else:
            total += min_payoffs[i-1]
    min_payoff = total / len(die_outcomes)
    min_payoffs[i] = min_payoff
    xvals[i] = i
    yvals[i] = int(min_payoff) + 1

plt.plot(np.array(xvals), np.array(min_payoffs), label="Theoretical Minimum")
plt.plot(np.array(xvals), np.array(yvals), label="Practical Minimum")
plt.title("Minimum acceptable roll by # of rolls remaing")
plt.xlabel("Number of rolls remaining")
plt.ylabel("Minimum lock-in roll")
plt.legend()
plt.show()
```
